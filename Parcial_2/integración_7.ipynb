```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

x = np.linspace(-1,1,1000)
y= np.linspace(-1,1,1000)

def esfera(x,y):
    return (1-x**2-y**2)**(1/2)

x1, y1 = np.meshgrid(x,y)
z1 = esfera(x1,y1)

axes = plt.axes(projection='3d')
axes.plot_surface(x1, y1, z1)
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.show()
```

```python
def volumen(x,y):
    vol = 0
    for i in range(1000):
        """z = esfera(x[i],y[i])"""
        for j in range(1000):
            z = esfera(x[i][j],y[i][j])
            if z**2+y[i][j]**2+x[i][j]**2 <=1:
                vol +=z
    return vol

vol = volumen(x1,y1)
area = (x1[1][1]-x1[1][2])**2
area
print("el volumen es de: " + str(vol*area))
```
