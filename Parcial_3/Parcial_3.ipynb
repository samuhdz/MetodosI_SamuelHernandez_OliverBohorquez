```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
#Ejercicio 22
#Definamos el sistema de numeros complejos
x=sym.Symbol('x', real=True)
y=sym.Symbol('y', real=True)
z=x+sym.I*y 
def funcion (z):
    f=z**6+(729/100)*(z**2)-(729/100)
    f=f.expand()
    return (sym.re(f),sym.im(f))

def Jacobiano (f):
    n=np.shape(f)[0]
    J=sym.zeros(n,n)
    for i in range(2):
        for j in range(2):
            if j==0:
                J[i,j] = sym.diff(f[i],x,1)
            else:
                J[i,j] = sym.diff(f[i],y,1)
    return J

f=funcion(z)
jacobiano=Jacobiano(f)
inversa_jacobiano=jacobiano.inv(method='LU')
f_num=sym.lambdify([x,y],f,'numpy')
jaco_num=sym.lambdify([x,y],jacobiano,'numpy')

def NewtonR (f,j,r,itmax=1000,error=1e-7):
    e = 1
    it = 0
    
    while e>error and it<itmax:
        fn=f(r[0],r[1])
        jn=j(r[0],r[1])
        ijn=np.linalg.inv(jn)
        rf=r-np.dot(ijn,fn)
        e=np.linalg.norm(r-rf)
        r=rf
        it+=1
       
    return r
def GetAllRoots (f,j):
    x=np.linspace(0,7,300)
    y=np.linspace(0,7,300)
    raices=[]
    for i in range(300):
        for j in range(300):
            entrada=NewtonR(f_num,jaco_num, [x[i],y[j]])
            if entrada not in raices:
                raices.append(raices)
    return raices


#Ejercicio 22
A=np.array([[0.2,0.1,1.,1.,0.],[0.1,4.,-1.,1.,-1.],[1.,-1.,60.,0.,-2.],[1.,1.,0.,8.,4.],[0.,-1.,-2.,4.,700.]])
vector_aux=np.array([1.,2.,3.,4.,5.])

def norma_infinita (vector):
    for i in vector:
        mayor=abs(vector[0])
        if abs(i)>mayor:
            mayor=i
    return abs(mayor)


def descenso_conjugado (A,b,x,tolerancia=0.01):
    r=np.dot(A,x)-b
    p=-r
    i=0
    r_infinito=norma_infinita(r)
    while r_infinito>tolerancia :
        a=-(np.dot(np.transpose(r),p))/np.dot((np.dot(np.transpose(p),A)),p)
        x_k=x+(a*p)
        r_k=np.dot(A,x_k)-b
        betha=np.dot(np.dot(np.transpose(r_k),A),p)/np.dot(np.transpose(p),np.dot(A,p))
        p_k=-r_k+(betha*p)
        r_infinito=norma_infinita(r_k)
        r=r_k
        x=x_k
        p=p_k      
        
        i+=1
        print(i)
        print(a)
    return x

ola=descenso_conjugado(A,vector_aux,np.array([12.,3.,4.,5.,6.]))
ola

```
