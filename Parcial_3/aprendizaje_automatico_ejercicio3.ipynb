```python
# ejercicio 3:
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as spo

def f(x):
    return np.absolute(x-2)

x = np.linspace(-10,10,200)
y = f(x) 

plt.plot(x,y)

```

```python
x0 = 5.
result = spo.minimize( f, x0 )
result
```

```python
_x = result.x
_f = result.fun
plt.plot(x,y)
plt.scatter(_x,_f)
```
