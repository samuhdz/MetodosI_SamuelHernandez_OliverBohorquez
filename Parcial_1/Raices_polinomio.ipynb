```python
import numpy as np
import matplotlib.pyplot as plt

def Functio(x):
    return (3*x**5) + (5*x**4) - (x**3)

x = np.linspace(-10,10,50)
y = Functio(x)

plt.plot(x,y)
plt.axhline(y = 0, color = 'r')

def Derivada(f,x,h=1e-8):
    return (f(x+h)-f(x-h))/(2*h)

def GetNewtonMethod(f, df, xn, itmax=100, precision=1e-8):
    error = 1
    it = 0
    while error > precision and it < itmax:
        try:
            xn1 = xn - f(xn)/df(f, xn)
            error = np.abs(f(xn)/df(f,xn))

        except ZeroDivisionError:
            print('Division por cero')
        
        xn = xn1
        it += 1 
    
    if it == itmax:
        return False
    else: 
        return xn
print(GetNewtonMethod(Functio,Derivada,0.3))
evaluar_raices=np.array([-2,0,0.14,0.3])
def GetAllRoots(x, tolerancia=10):
    
    Roots = np.array([])

    for i in x:
        root = GetNewtonMethod(Functio, Derivada, i)
        if root != False:
            croot = np.round(root, tolerancia)
            if croot not in Roots:
                Roots = np.append(Roots, croot)
    
    Roots.sort()
    return Roots

Roots = GetAllRoots(evaluar_raices)
print(Roots)
#Da pocas raices ya que la funcion tiene 3 raices cercanas a cero y el metodo "explota"
```
