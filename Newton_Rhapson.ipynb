```python
#Mi propio intento para el Método Newton Rhapson
#Primero, quiero definir la función distancia que voy a derivar más tarde
import numpy as np
import matplotlib.pyplot as plt

def eje_x (inicio, fin,cantidad):
    return np.linspace(inicio, fin, cantidad)

x=eje_x(-1,1,100)

def GetDistance (x):
    return np.sqrt(x**2+(3*x+5)**2)





#Ahora, vamos a graficar la derivada de la función
h=0.001
def derivada_central (y,x,h):
    derivada= (y(x+h)-y(x-h))/2*h
    return derivada

ejex=eje_x(-10,10,100)
ejey=derivada_central(GetDistance, ejex, h)

plt.scatter(ejex,ejey)
#Procedemos a hacer el método newton rhapson


    



```
